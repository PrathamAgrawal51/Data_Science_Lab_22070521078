```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.DataFrame(
    {
        'color': ['red', 'blue', 'green', 'blue', 'red']
    }
)
print(data)
```

```python
encoder = LabelEncoder()
```

```python
data['color_encoded'] = encoder.fit_transform(data['color'])
```

```python
print(data)
```
